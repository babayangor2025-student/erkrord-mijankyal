```python
graph = {
    "A": ["B", "C"],
    "B": ["D"],
    "C": ["E"],
    "D": ["F"],
    "E": [],
    "F": []
}

visited = set()

def dfs(node):
    if node in visited:
        return
    visited.add(node)
    print(node, end=" ")

    for neighbor in graph[node]:
        dfs(neighbor)


dfs("A")
```
