```python

def orientation(p, q, r):

    val = (q[1] - p[1]) * (r[0] - q[0]) - (q[0] - p[0]) * (r[1] - q[1])
    if val == 0:
        return 0
    elif val > 0:
        return 1
    else:
        return 2


def graham_scan(points):

    p0 = min(points, key=lambda p: (p[1], p[0]))


    def polar_angle(p):
        from math import atan2
        return atan2(p[1] - p0[1], p[0] - p0[0])

    sorted_points = sorted(points, key=lambda p: (polar_angle(p), (p[0]-p0[0])**2 + (p[1]-p0[1])**2))


    stack = []
    for pt in sorted_points:
        while len(stack) > 1 and orientation(stack[-2], stack[-1], pt) != 2:
            stack.pop()
        stack.append(pt)

    return stack

points = [(0,0), (1,1), (2,2), (0,3), (3,0), (3,3)]
convex_hull = graham_scan(points)
print("Convex Hull points’ù", convex_hull)
```
