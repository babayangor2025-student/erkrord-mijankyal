```python

edges = [
    ('A', 'B', 1),
    ('B', 'C', 3),
    ('A', 'C', 10),
    ('C', 'D', 2),
    ('D', 'B', -6)
]


vertices = ['A', 'B', 'C', 'D']


source = 'A'


distance = {v: float('inf') for v in vertices}
distance[source] = 0


for i in range(len(vertices) - 1):
    for u, v, w in edges:
        if distance[u] + w < distance[v]:
            distance[v] = distance[u] + w


negative_cycle = False
for u, v, w in edges:
    if distance[u] + w < distance[v]:
        negative_cycle = True
        break

# Արդյունք
if negative_cycle:
    print("Graph have negative sycle։")
else:
    print("Shortest ways from start՝")
    for v in vertices:
        print(f"{source} -> {v} = {distance[v]}")
```
