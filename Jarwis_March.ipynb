```python
def orientation(p, q, r):
    val = (q[1] - p[1]) * (r[0] - q[0]) - \
          (q[0] - p[0]) * (r[1] - q[1])

    if val == 0:
        return 0
    elif val > 0:
        return 1
    else:
        return 2
```

```python
def jarvis_march(points):
    if len(points) < 3:
        return points

    hull = []


    leftmost = 0
    for i in range(1, len(points)):
        if points[i][0] < points[leftmost][0]:
            leftmost = i

    p = leftmost

    while True:
        hull.append(points[p])

        q = (p + 1) % len(points)

        for i in range(len(points)):
            if orientation(points[p], points[i], points[q]) == 2:
                q = i

        p = q

        if p == leftmost:
            break

    return hull
```

```python
points = [(0,3), (2,2), (1,1), (2,1),
          (3,0), (0,0), (3,3)]

print(jarvis_march(points))
```
