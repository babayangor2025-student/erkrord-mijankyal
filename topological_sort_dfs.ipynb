```python
def topological_sort_dfs(graph):
    visited = set()
    order = []

    def dfs(v):
        if v in visited:
            return
        visited.add(v)

        for u in graph[v]:
            dfs(u)

        order.append(v)

    for v in graph:
        if v not in visited:
            dfs(v)

    order.reverse()
    return order



graph = {
    "A": ["C"],
    "B": ["C"],
    "C": ["D"],
    "D": []
}

print("Topological order (DFS):", topological_sort_dfs(graph))
```
