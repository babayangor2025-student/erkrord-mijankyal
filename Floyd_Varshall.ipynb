```python

vertices = ['A', 'B', 'C', 'D']


edges = [
    ('A', 'B', 5),
    ('A', 'C', 9),
    ('B', 'C', 3),
    ('C', 'D', 4),
    ('D', 'B', -9)
]

dist = {u: {v: float('inf') for v in vertices} for u in vertices}


for v in vertices:
    dist[v][v] = 0


for u, v, w in edges:
    dist[u][v] = w


for k in vertices:
    for i in vertices:
        for j in vertices:

            if dist[i][k] + dist[k][j] < dist[i][j]:
                dist[i][j] = dist[i][k] + dist[k][j]


print("Shortest ways between nodes՝")
for i in vertices:
    for j in vertices:
        if dist[i][j] == float('inf'):
            print(f"{i} -> {j} = ∞")
        else:
            print(f"{i} -> {j} = {dist[i][j]}")
    print()
```
